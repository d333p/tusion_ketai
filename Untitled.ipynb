## Generating Dataset

```python
#triangle = 2 * np.pi / 3
#square = np.pi / 2
def fig(n, alpha=0, r=1, x=0, y=0):
    angle = 2 * np.pi / n
    a = r * np.array([[np.sin(i * angle + alpha), np.cos(i * angle + alpha)] for i in range(n)])
    b = np.array([[x,y]] * n)
    return a - b
```

```python
from sklearn.utils.extmath import cartesian
```

```python
def draw_fig(fig, n=100, with_noise=False):
    pixels = cartesian([np.linspace(-2, 2, n), np.linspace(-2, 2, n)])
    pairs = []
    for i in range(len(fig) - 1):
        pairs.append(fig[i:i + 2])
    pairs.append(np.append([fig[-1]], [fig[0]], axis=0))
    pairs = np.array(pairs)
    out = np.zeros((len(pixels), ))
    l = 0
    for pix in pixels.reshape((n*n, 1, 2)):
        sign = np.sign(det(pairs[0] - pix))
        in_ = True
        for k in range(1, len(pairs)):
            if sign != np.sign(det(pairs[k] - pix)): 
                in_ = False
        if in_:
            out[l] = 10
        else:
            out[l] = 0
        l += 1
    out = out.reshape((n, n))
    if with_noise:
        out += np.random.normal(size=out.shape, scale=0.5)
    return out
```

```python
m = 5000
for i in tqdm(range(m)):
    tri = draw_fig(fig(3, alpha=float(i)/m * np.pi * 2, r=0.5) * np.pi, n=100, with_noise=True)
    sq = draw_fig(fig(4, alpha=float(i)/m * np.pi * 2, r=1.5), n=100, with_noise=True)
    np.savetxt('data/tri/' + str(i), tri, fmt='%.5f')
    np.savetxt('data/sq/' + str(i), sq, fmt='%.5f')
```

```python
m = 1000
for i in tqdm(range(m)):
    tri = draw_fig(fig(3, alpha=float(i)/m * np.pi * 2 / 10, r=0.5) * np.pi, n=100, with_noise=True)
    sq = draw_fig(fig(4, alpha=float(i)/m * np.pi * 2 / 10, r=1.5), n=100, with_noise=True)
    np.savetxt('data_noiseless/tri/' + str(i), tri, fmt='%.5f')
    np.savetxt('data_noiseless/sq/' + str(i), sq, fmt='%.5f')
```

## Building model

```python
m=1000
def read_data(n=m):
    tri_path = 'data_noiseless/tri/'
    sq_path = 'data_noiseless/sq/'
    tris = []
    sqs = []
    for i in tqdm(range(n)):
        tris.append(np.loadtxt(tri_path + str(i)))
        sqs.append(np.loadtxt(sq_path + str(i)))
    return np.stack(tris), np.stack(sqs)
```

```python
X, y = read_data()
```

```python
model = Sequential()
model.add(Convolution2D(16, 32, 32, input_shape=(100, 100, 1), activation='tanh'))
model.add(AveragePooling2D((2,2)))
model.add(Convolution2D(16, 8, 8, activation='tanh'))
model.add(Flatten())
model.add(Dense(int(model.output_shape[1]/16), activation='sigmoid'))
model.add(Reshape(target_shape=(27, 27, 1)))
model.add(UpSampling2D(size=(3,3)))
model.add(Convolution2D(16, 30, 30, activation='tanh'))
model.add(UpSampling2D(size=(2,2)))
model.add(Convolution2D(1, 5, 5, activation='linear'))
model.compile(loss='mse', optimizer='adam')
```

```python
train = X.reshape((-1, 100, 100, 1))
```

```python
y_train = y.reshape((-1, 100, 100, 1))
```

```python
model.fit(train, y_train, batch_size=100, nb_epoch=15)
```
